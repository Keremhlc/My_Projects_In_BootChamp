```python
no_null = kill[kill.isnull().sum(axis=1) == 0]
no_null.shape
```

## 1a: Box plot of age in both genders separeted by manner of death.

```python
kill.head()
```

```python
kill.manner_of_death.unique()
```

**INSTRUCTION: Demonstrate the Qurtiles of age in both genders separeted by manner of death.**

```python
# with seaborn
sns.boxplot(x = "gender", y = "age", data = kill, hue = "manner_of_death")
plt.show()
```

## 1b: Swarm Plot 

**INSTRUCTION: Demonstrate the age in both genders separeted by manner of death by a swarm plot.**

```python
# with seaborn
plt.figure(figsize=(11,6))
sns.swarmplot(x = "gender", y = "age", data = kill, hue = "manner_of_death")

plt.show()
```

## 2: Plot 12 most dangerous cities

```python
a = no_null.city.value_counts().head(12)
a
```

```python
# with seaborn barplot
plt.figure(figsize=(9,6))
plt.title("Most Dangerous Cities", c = "b", size=15)

a_num = range(len(a))
plt.xticks(a_num, a.index, rotation = 45)

sns.barplot(x= a.index, y = a.values)

plt.show()
```

```python
# with seaborn countplot
plt.figure(figsize=(8,6))
plt.title("Most Dangerous Cities", c = "b", size=15)
plt.xticks(rotation=45)

sns.countplot(x = "city", data=kill, order = a. index)

plt.show()
```

## 3: Plot 20 most dangerous states

```python
mds = no_null.state.value_counts().head(20)
mds
```

```python
# with seaborn barplot
plt.figure(figsize=(8,6))
sns.barplot(x = mds.index, y = mds.values, data=kill)

plt.title("Most Dangerous States", c="b", size=15)
plt.show()
```

```python
# with seaborn countplot
plt.figure(figsize=(8,6))
plt.title("Most Dangerous States", c = "b", size=15)

sns.countplot(x = "state", data = kill, order = mds.index)

plt.show()
```
