## Splitting the dataset into X(independent variables) and y (dependent variable)

```python
# y_pred = b1X + b0
#b1=slope , b0=Intecept
# Tahminler için Prediction ismini kullanıyoruz.
```

```python
X= df["total_spend"]
y= df["sales"]
```

## Determination of coefficients (slope and intercept)

```python
np.polyfit(X, y, deg=1)  # deg=1 Linear Regression olduğunu gösterir.
```

```python
slope, intercept = np.polyfit(X, y, deg=1)
```

```python
print("slope    :", slope) #b1
print("intercept:", intercept) #b0
```

```python
# Şöyle bi yorumda bulunabiliriz, Reklam harcaması yapmadan yapacağımız satış miktarı.
```

## Why do we use the least squares error method to find the regression line that best fits the data?

```python
b1, b0 = np.polyfit(X, y, deg=1)
```

```python
print("b1 :", b1)
print("b0 :", b0)
```

```python
y_pred = b1*X + b0 # b1 ve b0'ı bulduktan sonra, y_pred Değerini buluyoruz.
```

```python
# Residual = Greçek Değer - Tahmin Edilen Değer.
# LSE = Residual**2
values = {"actual": y, "predicted": y_pred, "residual":y-y_pred, "LSE": (y-y_pred)**2}
df_2 = pd.DataFrame(values)
df_2
```

```python
df_2.residual.sum().round() # Residuallarımızın toplamı sıfır(0) a eşit. Ondan dolayı LSE kullanıyoruz.
```

```python
df_2.LSE.sum()
```

## Prediction with simple linear regression

```python
predicted_sales_lin = b1* potential_spend + b0
predicted_sales_lin
```
