## Train | Test Split and Scaling

```python
X = df.drop('species',axis=1)
y = df['species']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=101)
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train) ## Fit sadece train'e.
X_test_scaled = scaler.transform(X_test)
```

## Modelling and Model Performance

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, plot_confusion_matrix
```

### With Default Parameters

```python
log_model = LogisticRegression()
```

```python
log_model.fit(X_train_scaled,y_train)
y_pred = log_model.predict(X_test_scaled)
y_pred ## sklearn kütüphanesi sadece target sütunu için,
        # object olarak kalmasına izin veriyor numerik değerlere çevirmeye gerek yok.

## Şunu unutmamalı ki katagorik verilere numara verip daha sonrada featuredaki correlasyonlara bakılamaz,
 # Ancak aralarında bir hiyerarşi varsa bakılabilir(ucuz-orta-pahalı) gibi.
```

```python
plot_confusion_matrix(log_model, X_test_scaled, y_test)

## Burada TP FN değerlerine bakabilmek için bir değere 1 verip diğre ikisine 0 veririz.
 # Sonrasında ise 1e1 T 0a0T 1e0 ve 0a1 F olur ve 1 içeren sütün ve satırlar da P değeri diğerleri ise N değeri alır.
```

```python
def eval_metric(model, X_train, y_train, X_test, y_test):
    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    
    print("Test_Set")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print()
    print("Train_Set")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
    
```

```python
eval_metric(log_model, X_train_scaled, y_train, X_test_scaled, y_test)

## Eğer veri setimizdeki 1 değer dengesizse(örn. versicolor 0.7) macro avg ye bakıyoruz.
 # Eğer dengesiz dediğimiz versicolora baktık fakat bizim için versicolor çok önemli değil ise,
   # yani çoğunluk sınıfının skorları daha önemli olduğunu söylüyorsak weighted avg ye bakabiliriz.
 # Eğer buradaki gibi dengeliyse hepsi accuracy'e bakabiliriz.
```

### Cross Validate

```python
from sklearn.model_selection import cross_validate

model = LogisticRegression()

scores = cross_validate(model, X_train_scaled, y_train, scoring = ['accuracy', 'precision_weighted','recall_weighted',
                                                                   'f1_weighted'], cv = 10)
                                                        ## Datamız binary olmadığı için, yani 0 ve 1 lerden oluşmadığı için,
                                                         # burada önceki notebooktaki(ML-9.1) gibi precision vb. yazamıyoruz.

df_scores = pd.DataFrame(scores, index = range(1, 11))
df_scores.mean()[2:]
```

### Cross Validate for versicolor

```python
from sklearn.metrics import make_scorer
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score

f1_versicolor = make_scorer(f1_score, average = None, labels =["versicolor"]) ## Binary durumda poslabels, multiclass'ta labels.
precision_versicolor = make_scorer(precision_score, average = None, labels =["versicolor"])
recall_versicolor = make_scorer(recall_score, average = None, labels =["versicolor"])

## ["Versicolor", "virginica"] şeklinde de yazabilirdik,
 # Ozaman average kısmını değiştirip istediğimiz(macro-weighted) bir şey yazardık.
 # Burada tek bir feature'ye bakacağımız için average'yi none olarak bırakabiliriz.
```

```python
model = LogisticRegression()

scores = cross_validate(model, X_train_scaled, y_train, scoring = {"f1_versicolor":f1_versicolor, 
                                                                   "precision_versicolor":precision_versicolor,
                                                                   "recall_versicolor":recall_versicolor}, cv = 10)
df_scores = pd.DataFrame(scores, index = range(1, 11))
df_scores.mean()[2:]


## Aslında olan değerlerim;
```

## Class prediction

```python
y_pred=log_model.predict(X_test_scaled)
y_pred_proba = log_model.predict_proba(X_test_scaled)

test_data = pd.concat([X_test, y_test], axis=1)
test_data["pred"] = y_pred
test_data["pred_proba_setosa"] = y_pred_proba[:,0]
test_data["pred_proba_versicolar"] = y_pred_proba[:,1]
test_data["pred_proba_virginica"] = y_pred_proba[:,2]
test_data.sample(10)

## Burada 0.5 gibi bir durum yok, burada 0 ila 1 arasında 3 olasılığa dağıtıyor değerleri,
 # Ve en büyük olasılık hangisiyse os sınıfa atıyor. Multiclass'ta.
```

### With Best Parameters (GridsearchCV)

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html

```python
log_model = LogisticRegression(max_iter=5000)
```

```python
penalty = ["l1", "l2"]
C = np.logspace(-1, 5, 20)

param_grid = {"penalty" : penalty,
             "C" : C}
```

```python
## Gridsearch'de scoring yapmassak default olarak accuracy değeri döndürür.
 # Datamızda burada dengeli olduğu için farklı bir scor üretmedik.   

grid_model = GridSearchCV(log_model, param_grid = param_grid, cv=5) 
#scoring = f1_setosa = make_scorer(f1_score, average = None, labels =["setosa"])
```

```python
grid_model.fit(X_train_scaled,y_train)
```

```python
grid_model.best_params_
```

```python
grid_model.best_score_
```

```python
y_pred = grid_model.predict(X_test_scaled)
y_pred
```

```python
plot_confusion_matrix(grid_model, X_test_scaled, y_test)

## Gridsearch sonrasında tekrardan prediction yaptırdım, 1 hatam vardı(18) o da gitti.
```

```python
eval_metric(grid_model, X_train_scaled, y_train, X_test_scaled, y_test)

## Train setimizde 2 adet hata var, onun haricinde test setimizdeki scorların gayet dengeli olduğunu görüyoruz.
```

## ROC (Receiver Operating Curve) and AUC (Area Under Curve)

```python
from sklearn.metrics import plot_roc_curve

## roc-auc score sadece binary modeller için çizdirilebilir.
 # Bu durumu sklearn şöyle çözmüş, 3 sınıfımızın 1'ini 1 yapıp diğerlerini 0 yapıyor.
 # Ve bu işlemi her değer'e 1 gelecek şekilde tekrarlıyor. (37)
```

```python
plot_roc_curve(grid_model, X_test_scaled, y_test);
```

https://scikit-learn.org/stable/auto_examples/model_selection/plot_roc.html

```python
from yellowbrick.classifier import ROCAUC
model = grid_model
visualizer = ROCAUC(model)

visualizer.fit(X_train_scaled, y_train)        # Fit the training data to the visualizer
visualizer.score(X_test_scaled, y_test)        # Evaluate the model on the test data
visualizer.show();

## Tüm sınıflarım için(versicolor-setosa-vb.) 1 değeri dönmüş,
 # Buda modelimizin bu sınıfları çok güzel bir şekilde tahmin ettiğini gösteriyor.
```

```python
from yellowbrick.classifier import PrecisionRecallCurve

model = grid_model

viz = PrecisionRecallCurve(
    model,
    per_class=True,  ## Bunu eğer True yapmazsak binarymiiş gibi hesaplar.
    cmap="Set1"
)
viz.fit(X_train_scaled, y_train)
viz.score(X_test_scaled, y_test)
viz.show();

## Mesela buradan şunu anlıyoruz ki dengeli datalarda precision-recall kullanılmalı.
 # Çünkü, ROC-AUC da tüm skorlarımız iyi olarak görünüyor fakat precision-recall da bir skorumuz 0.69 çıktı.
```
